### 2. Построить обзор ключевых метрик

На этом этапе нам нужно выполнить верхнеуровневую оценку состояния продукта с помощью собранных витрин. Бизнес просит просто оценить текущее состояние — без глубокой сегментации и оценки сложных метрик.

Выручка бизнеса — 5% комиссии от стоимости любой продажи.

Посмотрим на общую динамику ключевых метрик

Revenue

````python
revenue_month = pa_marketplace_orders.groupby('order_month')['total_price'].sum().reset_index()
revenue_month['Revenue'] = revenue_month['total_price'] * 0.05

plt.figure(figsize=(12, 6))
sns.lineplot(data=revenue_month, x='order_month', y='Revenue', marker='o')
plt.title('Revenue по месяцам')
plt.ylabel('Revenue')
plt.xlabel('Месяц')
plt.show()
````

В начале года показатели выручки низкие (меньше 0.75). Однако с течением года выручка постепенно растет в первые месяцы года, достигая пика примерно в май–август (2.50). После августа наблюдается снижение выручки, особенно заметное в последние месяцы года, где сумма выручки опустилась до отметки 1.25.

Рост в первой половине года может быть связан с активным привлечением новых пользователей, сезонными акциями или успешными маркетинговыми кампаниями. Снижение во второй половине года может указывать на снижение активности пользователей, сезонность спроса на товары, недостаточное удержание клиентов.

AOV

````python
aov_month = pa_marketplace_orders.groupby('order_month').agg(
    total_price_sum=('total_price','sum'),
    orders=('order_id','count')
).reset_index()
aov_month['AOV'] = aov_month['total_price_sum'] / aov_month['orders']

plt.figure(figsize=(12, 6))
sns.lineplot(data=aov_month, x='order_month', y='AOV', marker='o')
plt.title('AOV по месяцам')
plt.ylabel('Средний чек (AOV)')
plt.xlabel('Месяц')
plt.show()
````

В начале года (январь–февраль) средний чек относительно низкий (13450–13550). С марта начинается рост, с небольшими колебаниями до сентября (13800–14500). Наибольший всплеск наблюдается в ноябре -  около 14850. В декабре наблюдается резкое падение обратно до 13700.

Рост с марта по август–сентябрь- это, возможно, эффект сезонных покупок, акций или маркетинговых кампаний, пользователи совершают больше крупных заказов. Пик в ноябре возможно связан с ежегодной черной пятницей, средний чек резко увеличивается за счет крупных заказов. Снижение в декабре, возможно, связано с тем, что пользователи уже купили подарки на новый год в ноябре на скидках, а январские праздники, где люди тоже часто тратят деньги, еще не наступили

ARPU и ARPPU

````python
users_month = pa_marketplace_users.groupby('cohort_month')['user_id'].nunique().reset_index(name='users')

paying_users_month = pa_marketplace_orders.groupby('order_month')['user_id'].nunique().reset_index(name='paying_users')

arpu_df = revenue_month.merge(users_month, left_on='order_month', right_on='cohort_month', how='left')
arpu_df = arpu_df.merge(paying_users_month, on='order_month', how='left')
arpu_df['ARPU'] = arpu_df['Revenue'] / arpu_df['users']
arpu_df['ARPPU'] = arpu_df['Revenue'] / arpu_df['paying_users']

plt.figure(figsize=(12, 6))
sns.lineplot(data=arpu_df, x='order_month', y='ARPU', marker='o', label='ARPU')
sns.lineplot(data=arpu_df, x='order_month', y='ARPPU', marker='o', label='ARPPU')
plt.title('ARPU и ARPPU по месяцам')
plt.ylabel('Доход на пользователя')
plt.xlabel('Месяц')
plt.legend()
plt.show()
````

ARPU

В январе очень низкий , примерно 220, постепенно растет до мая примерно 800. После мая наблюдается постепенное снижение до конца года примерно 300 в декабре.

ARPPU

В начале года около 1430, постепенно растет, в августе наступает пик - 1780. После пика летом наблюдается умеренное снижение и колебания. В конце года значение опускается до примерно 1650.

Разрыв между ARPU и ARPPU большой, что говорит о низкой конверсии пользователей в платящих пользователей. ARPU растет только в первые месяцы года, затем снижается, в то время как ARPPU держится на высоком уровне. Это значит, что доход с платящих пользователей стабильный, но большинство пользователей не платят, что снижает средний доход на всех пользователей.

DAU, WAU, MAU

````python
# DAU
dau = pa_marketplace_events.groupby('event_date')['user_id'].nunique().reset_index()
dau.rename(columns={'user_id': 'DAU'}, inplace=True)

plt.figure(figsize=(12,6))
sns.lineplot(data=dau, x='event_date', y='DAU', marker='o')
plt.title('DAU')
plt.xlabel('Дата')
plt.ylabel('Количество уникальных пользователей')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# WAU
wau = pa_marketplace_events.groupby('event_week')['user_id'].nunique().reset_index()
wau.rename(columns={'user_id': 'WAU'}, inplace=True)

plt.figure(figsize=(12,6))
sns.lineplot(data=wau, x='event_week', y='WAU', marker='o', color='orange')
plt.title('WAU')
plt.xlabel('Неделя')
plt.ylabel('Количество уникальных пользователей')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


# MAU
mau = pa_marketplace_events.groupby('event_month')['user_id'].nunique().reset_index()
mau.rename(columns={'user_id': 'MAU'}, inplace=True)

plt.figure(figsize=(12,6))
sns.lineplot(data=mau, x='event_month', y='MAU', marker='o', color='green')
plt.title('MAU')
plt.xlabel('Месяц')
plt.ylabel('Количество уникальных пользователей')
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

DAU

В январе практически нулевая активность. С февраля начинается постепенный рост, достигая в мае–июне примерно 430 пользователей. Пик активности приходится на сентябрь - около 450 пользователей. После сентября наблюдается снижение, к декабрю DAU опускается до 300. Это говорит о том, что ежедневная активность пользователей постепенно растет, но большинство пользователей не заходят на платформу каждый день.

WAU

В январе около 100 пользователей, с февраля начинается активный рост, в августе–сентябре достигает максимума примерно 1450 пользователей. После сентября начинается снижение до 1300 в ноябре и резкий спад до 300 в декабре. CTR

````python
clicks = pa_marketplace_events[pa_marketplace_events['event_type'].isin(['product_click', 'banner_click'])].shape[0]
views = pa_marketplace_events[pa_marketplace_events['event_type'].isin(['product_view', 'banner_click', 'promo_view'])].shape[0]

CTR = clicks / views * 100

pa_marketplace_events['event_month'] = pa_marketplace_events['event_date'].dt.to_period('M').apply(lambda r: r.start_time)

monthly_events = pa_marketplace_events.groupby('event_month')['event_type'].value_counts().unstack(fill_value=0)
monthly_events['CTR'] = (monthly_events['product_click'] + monthly_events['banner_click']) / \
                        (monthly_events['product_view'] + monthly_events['banner_click'] + monthly_events['promo_view']) * 100

plt.figure(figsize=(12, 6))
sns.lineplot(data=monthly_events, x=monthly_events.index, y='CTR', marker='o')
plt.title('CTR по месяцам')
plt.xlabel('Месяц')
plt.ylabel('CTR')
plt.xticks(rotation=45)
plt.show()
````

В январе CTR был высокий, около 88.8%. В феврале наблюдается небольшой рост до 90%, затем с марта по август показатель стабильно держится в диапазоне 88.85–89.15%. В сентябре начинается постепенное снижение CTR до конца года, в сентябре 88.35%, в октябре 87.75%, в ноябре 88.4%, а в декабре достигает минимума - 87.4%.

CTR демонстрирует стабильную высокую кликабельность пользователей на протяжении большей части года, что говорит о заинтересованности аудитории в товарах. Снижение во второй половине года хоть и незначительное, но заметное, и может потребовать анализа.

CAC

````python
all_users = pa_marketplace_users.copy()
all_users['month'] = all_users['registration_date'].dt.to_period('M').dt.to_timestamp()

monthly_new_users = all_users.groupby('month').agg(new_users=('user_id', 'count')).reset_index()
monthly_marketing = pa_marketplace_campaign_costs.groupby('spend_month').agg(marketing_spend=('budget','sum')).reset_index()

cac_monthly = monthly_new_users.merge(monthly_marketing, left_on='month', right_on='spend_month', how='left')
cac_monthly['CAC'] = cac_monthly['marketing_spend'] / cac_monthly['new_users']

plt.figure(figsize=(12,5))
plt.plot(cac_monthly['month'], cac_monthly['CAC'], marker='o', linestyle='-')
plt.title('CAC по месяцам')
plt.xlabel('Месяц')
plt.ylabel('CAC, руб.')
plt.xticks(rotation=45)
plt.show()
````

В мае и июне показатель продолжает рост, достигая 400 и 402,5. В июле и августе стоимость стабилизируется на более высоком уровне - 405–408,5. С сентября начинается небольшое снижение до конца года, в сентябре 407,5, в октябре 408, в ноябре 405, а в декабре 404,5.

Во второй половине года наблюдается стабилизация и постепенное снижение CAC, что говорит о более эффективном управлении рекламным бюджетом, но CAC все равно остается на высоком уровне.

ROI

````python
roi_df = revenue_month.merge(
    cac_monthly[['month', 'marketing_spend']],
    left_on='order_month',
    right_on='month',
    how='left'
)

roi_df['ROI'] = (roi_df['Revenue'] - roi_df['marketing_spend']) / roi_df['marketing_spend']

plt.figure(figsize=(12, 6))
sns.lineplot(data=roi_df, x='order_month', y='ROI', marker='o')
plt.axhline(0, linestyle='--', color='gray')

plt.title('ROI по месяцам')
plt.xlabel('Месяц')
plt.ylabel('ROI')
plt.show()
````

В январе ROI был отрицательный, около -0,4, что указывает на то, что затраты на маркетинг превышали доходы. Payback Period

````python
monthly_revenue_per_user = revenue_month.merge(
    monthly_new_users, left_on='order_month', right_on='month', how='left'
)
monthly_revenue_per_user['Revenue_per_user'] = monthly_revenue_per_user['Revenue'] / monthly_revenue_per_user['new_users']

payback_df = cac_monthly.merge(
    monthly_revenue_per_user[['month', 'Revenue_per_user']],
    left_on='month',
    right_on='month',
    how='left'
)
payback_df['Payback_Months'] = payback_df['CAC'] / payback_df['Revenue_per_user']

plt.figure(figsize=(12,6))
sns.lineplot(data=payback_df, x='month', y='Payback_Months', marker='o')
plt.title('Payback Period по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Payback Period (мес)')
plt.xticks(rotation=45)
plt.show()
````

В январе Payback Period был достаточно высоким - 1.8 месяца, что говорит о том, что маркетплейсу требовалось почти два месяца, чтобы окупить затраты на привлечение одного пользователя. Пик эффективности приходится на апрель–май, когда Payback Period составляет около 0.5 месяца, то есть доход с пользователя быстро покрывает CAC. В июне показатель немного увеличивается до 0.55 месяца, после чего в июле и августе стабилизируется на уровне 0.8 месяца. С сентября начинается постепенный рост Payback Period, достигая 1.4 месяца в декабре. Это говорит о замедлении окупаемости пользователей в конце года.

Таким образом, график демонстрирует значительное улучшение окупаемости в первые месяцы года, что, вероятно, связано с ростом ARPU или корректно выверенном CAC. Пик эффективности - апрель–май. Начиная с июня, эффективность постепенно снижается, и к декабрю период окупаемости возвращается почти к уровню начала года. Средний чек увеличивается к ноябрю благодаря крупным сезонным покупкам, но резко снижается в декабре. Период окупаемости увеличивается, рентабельность падает, стоимость привлечения пользователя остаётся высокой. Основные проблемы второй половины года - низкая конверсия обычных пользователей в платящих, нерегулярная активность пользователей, замедление окупаемости и снижение рентабельности. Для поддержания устойчивой доходности необходимы меры по удержанию пользователей, стимулированию активности, повышению конверсии и оптимизации маркетинговых расходов на протяжении всего года.

````python
pa_marketplace_users['reg_month'] = pa_marketplace_users['registration_date'].dt.to_period('M').dt.to_timestamp()

monthly_marketing = pa_marketplace_campaign_costs.groupby('spend_month').agg(marketing_spend=('budget','sum')).reset_index()
monthly_users = pa_marketplace_users.groupby('reg_month').agg(new_users=('user_id','count')).reset_index()
cac_df = monthly_users.merge(monthly_marketing, left_on='reg_month', right_on='spend_month', how='left')
cac_df['CAC'] = cac_df['marketing_spend']

orders = pa_marketplace_orders.copy()
orders = orders.merge(pa_marketplace_users[['user_id','reg_month']], on='user_id', how='left')

orders['N_month'] = ((orders['order_month'] - orders['reg_month']) / np.timedelta64(1,'M')).round(0)
orders = orders[np.isfinite(orders['N_month'])]

orders['revenue'] = orders['total_price'] * 0.05

orders = orders.sort_values(['reg_month','N_month'])
cohort_revenue = orders.groupby(['reg_month','N_month']).agg(monthly_revenue=('revenue','sum')).reset_index()
cohort_revenue = cohort_revenue.merge(cac_df[['reg_month','CAC']], on='reg_month', how='left')
cohort_revenue['cum_revenue'] = cohort_revenue.groupby('reg_month')['monthly_revenue'].cumsum()

cohort_revenue['ROI'] = cohort_revenue['cum_revenue'] / cohort_revenue['CAC'] - 1
cohort_revenue['ROI'] = cohort_revenue['ROI'].round(2)

roi_pivot = cohort_revenue.pivot(index='reg_month', columns='N_month', values='ROI')

roi_pivot.index = roi_pivot.index.strftime('%Y-%m')

plt.figure(figsize=(12, 8))
sns.heatmap(
    roi_pivot,
    annot=True,
    fmt=".2f",
    cmap="RdYlGn",
    center=0,
    cbar_kws={'label': 'ROI'}
)
plt.title("Когортный ROI по месяцам после регистрации")
plt.xlabel("Месяцы после регистрации")
plt.ylabel("Когорта")
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
````

Ранние когорты достигают положительного ROI на 2 месяце жизни, в то время как когорты второй половины года могут оставаться убыточными даже на горизонте 3–4 месяцев. Это указывает на ухудшение качества привлечённого трафика и необходимость пересмотра каналов и бюджетов во второй половине года.

````python
funnel_steps = [
    'page_view',
    'product_view',
    'product_click',
    'add_to_cart',
    'checkout_start',
    'checkout_complete'
]

funnel_events = pa_marketplace_events[pa_marketplace_events['event_type'].isin(funnel_steps)]

funnel_counts = (
    funnel_events.groupby('event_type')['user_id']
    .nunique()
    .reindex(funnel_steps)
)

funnel_df = pd.DataFrame({
    'step': funnel_counts.index,
    'users': funnel_counts.values
})

funnel_df['conversion_from_first_%'] = (
    funnel_df['users'] / funnel_df['users'].iloc[0] * 100
).round(1)

funnel_df['conversion_from_previous_%'] = (
    funnel_df['users'] / funnel_df['users'].shift(1) * 100
).round(1)



plt.figure(figsize=(10, 6))
plt.bar(
    funnel_df['step'][1:], 
    funnel_df['conversion_from_previous_%'][1:],
    color='skyblue'
)
plt.title('Пошаговая воронка: конверсия от предыдущего шага')
plt.ylabel('Конверсия от предыдущего шага (%)')
plt.xlabel('Этапы воронки')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.show()



plt.figure(figsize=(10, 6))
plt.bar(
    funnel_df['step'],
    funnel_df['conversion_from_first_%'],
    color='lightgreen'
)
plt.title('Классическая воронка: доля пользователей от первого шага')
plt.ylabel('Доля пользователей (%)')
plt.xlabel('Этапы воронки')
plt.xticks(rotation=45)
plt.ylim(0, 100)
plt.show()
````

Интересно отдельно изучить category_name, region, os, device по месяцам

````python
pa_marketplace_orders['revenue'] = pa_marketplace_orders['total_price'] * 0.05

orders_users = pa_marketplace_orders.merge(
    pa_marketplace_users[['user_id', 'region']],
    on='user_id', how='left'
)

last_event = pa_marketplace_events.sort_values('event_date').drop_duplicates('user_id', keep='last')
orders_users = orders_users.merge(
    last_event[['user_id', 'os', 'device']],
    on='user_id', how='left'
)

def plot_monthly_metric(df, group_col, metric_col='revenue', agg_func='sum', title=''):
    monthly = df.groupby([group_col, 'order_month']).agg(
        metric=(metric_col, agg_func)
    ).reset_index()
    
    plt.figure(figsize=(14,6))
    sns.lineplot(data=monthly, x='order_month', y='metric', hue=group_col, marker='o')
    plt.title(title)
    plt.ylabel(metric_col.capitalize())
    plt.xlabel('Месяц')
    plt.xticks(rotation=45)
    plt.show()

#  Анализ по category_name

top5 = (
    orders_users
    .groupby('category_name')['revenue']
    .sum()
    .sort_values(ascending=False)
    .head(5)
    .index
)

plt.figure(figsize=(14,6))

for name, data in orders_users.groupby('category_name'):
    monthly = (
        data
        .groupby('order_month')['revenue']
        .sum()
        .reset_index()
        .sort_values('order_month')
    )
    
    if name in top5:
        plt.plot(monthly['order_month'], monthly['revenue'], label=name)
    else:
        plt.plot(monthly['order_month'], monthly['revenue'], alpha=0.3)

plt.title('Выручка по категориям товаров по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Выручка')
plt.legend(title='Категории (топ-5)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()


#  Анализ по region
plot_monthly_metric(
    orders_users, 
    group_col='region', 
    metric_col='revenue', 
    agg_func='sum', 
    title='Выручка по регионам по месяцам'
)

# Анализ по os
plot_monthly_metric(
    orders_users, 
    group_col='os', 
    metric_col='revenue', 
    agg_func='sum', 
    title='Выручка по операционным системам по месяцам'
)

# Анализ по device
plot_monthly_metric(
    orders_users, 
    group_col='device', 
    metric_col='revenue', 
    agg_func='sum', 
    title='Выручка по типу устройства по месяцам'
)
````

# Дизайн эксперимента:

Бонус на первую покупку для пользователей TikTok

## 1. Контекст и цель эксперимента

- **Бизнес-контекст:** Команда продукта запускает эксперимент, направленный на повышение конверсии среди пользователей, привлечённых через канал TikTok. Для стимулирования первой покупки пользователям будет предоставлен бонус на первую покупку.  
- **Цель:** Проверить, влияет ли предоставление бонуса на первую покупку на конверсию и выручку среди пользователей TikTok.  
- **Гипотеза:** Предоставление бонуса на первую покупку новым пользователям, привлеченным через TikTok, приведет к увеличению конверсии в первую покупку

---

## 2. Формат эксперимента

- **Тип эксперимента:** Сплит-роллинг (rolling split)  
- **Канал:** Только пользователи, пришедшие с TikTok  
- **Группы:**  
  - **Контрольная (A):** пользователи TikTok без бонуса  
  - **Тестовая (B):** пользователи TikTok с бонусом на первую покупку  
- **Распределение:** Равномерное распределение новых пользователей TikTok между группами по дате регистрации  
- **Период проведения:** 3 недели (фиксировано)

---

## 3. Ключевые метрики

### Основная метрика

- **Конверсия в первую покупку** — доля пользователей, зарегистрировавшихся с TikTok и совершивших первую покупку  
  - *Цель:* увеличить конверсию с помощью бонуса  
  - *Критерий успеха:* статистически значимый рост конверсии в тестовой группе

### Вспомогательные метрики

- **ARPU** (средняя выручка на пользователя TikTok)  
- **ARPPU** (средняя выручка на покупателя TikTok)  
- **AOV** (средний чек TikTok)

### Барьерная метрика

- **Общая выручка от пользователей TikTok**

---

## 4. Размеры групп и запуск

- **Целевая аудитория:** только пользователи TikTok  
- **Размеры групп:** определяются по фактическому трафику TikTok за период эксперимента  
- **Комментарий:** запуск без предварительного расчёта мощности и MDE из-за срочности задачи

Проверим корректность распределения новых пользователей по группам А/В-эксперимента.

````python
user_orders = pa_diploma_orders_AB.groupby('user_id')['total_price'].sum().reset_index()

user_orders = pa_diploma_AB_split_users.merge(user_orders, on='user_id', how='left')

user_orders['total_price'] = user_orders['total_price'].fillna(0)


plt.figure(figsize=(10,5))

for grp in user_orders['group'].unique():
    subset = user_orders[user_orders['group'] == grp]
    plt.hist(subset['total_price'], bins=50, alpha=0.6, label=grp)

plt.title('Распределение суммы заказов на пользователя по группам A/B-теста')
plt.xlabel('Сумма заказов на пользователя')
plt.ylabel('Количество пользователей')
plt.legend()
plt.show()
````

Гистограмма распределения сумм заказов на пользователя по контрольной и тестовой группам показывает, что распределения ключевой метрики в обеих группах имеют схожую форму и характер. Барьерная метрика

Для каждой метрики также выведим:

значения в контрольной и тестовой группах,
абсолютные и относительные изменения значений в тестовой группе по сравнению с контрольной.

````python
tiktok_users = pa_diploma_AB_split_users.merge(
    pa_diploma_users_AB[['user_id', 'registration_date', 'acq_channel']],
    on='user_id',
    how='left'
)
tiktok_users = tiktok_users[tiktok_users['acq_channel'] == 'TikTok']

# Берем  3 недели
tiktok_users['registration_date'] = pd.to_datetime(tiktok_users['registration_date'])
tiktok_users = tiktok_users[
    (tiktok_users['registration_date'] >= '2025-01-06') &
    (tiktok_users['registration_date'] <= '2025-01-26')
]

tiktok_orders = pa_diploma_orders_AB.merge(
    tiktok_users[['user_id', 'group']],
    on='user_id',
    how='inner'
)

# Основная метрика. конверсия в покупку
first_orders = tiktok_orders.groupby('user_id')['order_date'].min().reset_index()
first_orders['converted'] = 1
conv_reg_to_first = tiktok_users.merge(first_orders[['user_id','converted']], on='user_id', how='left')
conv_reg_to_first['converted'] = conv_reg_to_first['converted'].fillna(0)

# Вспомогательные метрики. ARPU
arpu = tiktok_orders.groupby('user_id')['total_price'].sum().reset_index()
arpu = tiktok_users.merge(arpu, on='user_id', how='left').fillna(0)

# Вспомогательные метрики. ARPPU
arppu = tiktok_orders.groupby('user_id')['total_price'].sum().reset_index()
arppu = arppu.merge(tiktok_users[['user_id','group']], on='user_id', how='left')

# Вспомогательные метрики. AOV
aov = tiktok_orders.groupby('order_id')['total_price'].sum().reset_index()
aov = aov.merge(tiktok_orders[['order_id','user_id','group']].drop_duplicates(), on='order_id', how='left')

# Барьерная метрика. общая выручка от пользователей TikTok по группам
total_revenue = tiktok_orders.groupby('group')['total_price'].sum().reset_index()
total_revenue = total_revenue.set_index('group')['total_price']

metrics_table = pd.DataFrame({
    'metric': [
        'conv_reg_to_first_purchase',
        'ARPU',
        'ARPPU',
        'AOV',
        'total_revenue'
    ],
    'control': [
        conv_reg_to_first[conv_reg_to_first['group']=='control']['converted'].mean(),
        arpu[arpu['group']=='control']['total_price'].mean(),
        arppu[arppu['group']=='control']['total_price'].mean(),
        aov[aov['group']=='control']['total_price'].mean(),
        total_revenue['control']
    ],
    'test': [
        conv_reg_to_first[conv_reg_to_first['group']=='treatment']['converted'].mean(),
        arpu[arpu['group']=='treatment']['total_price'].mean(),
        arppu[arppu['group']=='treatment']['total_price'].mean(),
        aov[aov['group']=='treatment']['total_price'].mean(),
        total_revenue['treatment']
    ]
})

# Абсолютные изменения
metrics_table['abs_change_test_vs_control'] = metrics_table['test'] - metrics_table['control']

# Относительные изменения в %
metrics_table['rel_change_test_vs_control_%'] = metrics_table['abs_change_test_vs_control'] / metrics_table['control'] * 100

metrics_table[['control','test','abs_change_test_vs_control','rel_change_test_vs_control_%']] = \
    metrics_table[['control','test','abs_change_test_vs_control','rel_change_test_vs_control_%']].round(2)

metrics_table
````

Проверим статистическую значимость различий метрик между группами эксперимента.

Ключевая метрика конверсия в первую покупку. Когда мы анализируем конверси - это всегда z-test

H0: Конверсия в первую покупку у пользователей TikTok одинакова в контрольной и тестовой группах. Различия между группами отсутствуют

H1: Конверсия в первую покупку у пользователей TikTok различается между контрольной и тестовой группами.

````python
successes_control = conv_reg_to_first[conv_reg_to_first['group']=='control']['converted'].sum()
successes_test = conv_reg_to_first[conv_reg_to_first['group']=='treatment']['converted'].sum()

trials_control = len(conv_reg_to_first[conv_reg_to_first['group']=='control'])
trials_test = len(conv_reg_to_first[conv_reg_to_first['group']=='treatment'])

def z_test(successes1, successes2, trials1, trials2, alpha=0.05):
    p1 = successes1 / trials1
    p2 = successes2 / trials2
    p_combined = (successes1 + successes2) / (trials1 + trials2)
    difference = p1 - p2
    z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trials1 + 1/trials2))
    p_value = (1 - st.norm.cdf(abs(z_value))) * 2
    return p_value

p_value = z_test(successes_control, successes_test, trials_control, trials_test)
print("Ключевая метрика - конверсия в первую покупку")
print(f"p-value: {p_value:.4f}")
if p_value < 0.05:
    print("Отвергаем H0: различия статистически значимы\n")
else:
    print("Не получилось отвергнуть H0: различия незначимы\n")
````

Вспомогательная метрика средняя выручка на пользователя TikTok. Это числовая метрика, используем t-test

H0: Средняя выручка на пользователя TikTok одинакова в контрольной и тестовой группах. Предоставление бонуса не влияет на ARPU. 

H1: Средняя выручка на пользователя TikTok различается между контрольной и тестовой группами.

Цель эксперимента заключалась в проверке того, приводит ли бонус на первую покупку к росту конверсии и выручки без негативного влияния на метрики.


Анализ пользовательского пути

Воронки пользовательского пути показали, что поведение пользователей на этапах -просмотр страницы, клик по товару, добавление в корзину, было схожим в контрольной и тестовой группах. Конверсии на этих этапах высоки, более 93% для контрольной группы и почти 98% для тестовой. Это подтверждает корректное распределение пользователей между группами и отсутствие влияния бонуса на первоначальный интерес к продукту и взаимодействие с каталогом.

Различия начинаются на этапе оформления заказа, 65,9% пользователей тестовой группы, добавивших товар в корзину, переходят к оформлению заказа, против 59,6% в контрольной. На финальном этапе завершения покупки тестовая группа также показывает более высокую конверсию - 64,4% против 62,5% у контрольной. Итоговая конверсия от регистрации до покупки в тестовой группе составила 35,3%, против 29,8% у контрольной. Это подтверждает, что бонус стимулирует пользователей к совершению первой покупки, особенно на последних этапах пути.

Метрики и их анализ

1. Конверсия в первую покупку

В тестовой группе 7% новых пользователей TikTok совершили первую покупку, в контрольной - только 1%. Это значит, что бонус увеличил количество покупателей почти в 7 раз. Эффект статистически значимый, то есть с высокой уверенностью можно сказать, что рост произошёл из-за бонуса.

2. Средняя выручка на пользователя (ARPU)

В тестовой группе средняя выручка с каждого пользователя выросла с 213 до 2 253 единиц. Это показывает, что новые пользователи, привлечённые через TikTok и получившие бонус, стали приносить гораздо больше денег.

3. Средняя выручка на покупателя 

Среди тех, кто сделал покупку, средняя сумма выросла с 29 654 до 32 433 единиц. Прирост небольшой (около 9%), что говорит о том, что бонус больше привлекает новых покупателей, а средний чек на покупателя изменяется незначительно.

4. Средний чек

Средний чек снизился с 19 769 до 15 484 единиц. Новые пользователи, скорее всего, используют бонус для совершения более мелких покупок. Это не страшно, так как количество покупателей и конверсия выросли, что компенсирует снижение среднего чека.

5. Общая выручка от пользователей TikTok

Общая выручка в тестовой группе составила 2 400 069 единиц против 237 233 единиц в контрольной. Рост более чем в 10 раз. Это значит, что бонус не только увеличил количество покупателей, но и значительно поднял общую выручку.



Бонус на первую покупку существенно увеличивает конверсию и общую выручку, что подтверждает гипотезу о положительном влиянии финансовой стимуляции.

Снижение среднего чека и отсутствие значимых изменений в средняя выручка на покупателя указывают на то, что новые пользователи делают более мелкие покупки. При этом рост числа покупателей и конверсии полностью компенсирует этот эффект.

Ранние этапы пути, такие как взаимодействие с каталогом и корзиной, не изменились, что говорит о стабильности поведения пользователей и корректной сегментации.

Проблемы эксперимента

Во-первых, эксперимент запускался без предварительного расчёта мощности и минимально детектируемого эффекта. Это было осознанное решение из-за срочности задачи, однако оно повышает риск переоценки эффекта.

Во-вторых, эксперимент длился чуть больше трёх недель. Этого достаточно для оценки конверсии в первую покупку, но недостаточно для анализа повторных покупок, удержания и долгосрочной ценности пользователей.

В-третьих, наблюдается снижение среднего чека. Хотя оно компенсируется ростом конверсии и общей выручки, при масштабировании механики этот эффект может усилиться и потребует отдельного контроля.

В-четвертых,  эксперимент охватывал только канал TikTok, и полученные выводы нельзя автоматически переносить на другие каналы привлечения.

Рекомендации

1. Продолжить использование бонуса. Эксперимент показал значительный положительный эффект на конверсию и выручку, поэтому использование бонуса - это хорошо.

2. Мониторинг среднего чека. Снижение среднего чека требует контроля.

3. Продолжать наблюдение. Продолжить мониторинг поведения пользователей в течение нескольких недель, чтобы оценить удержание и повторные покупки.

Выводы

Эксперимент показал, что бонус на первую покупку является эффективным инструментом быстрого роста конверсии и выручки среди новых пользователей TikTok. Он не влияет на ранние этапы пользовательского пути, но существенно повышает вероятность завершения покупки на финальных этапах.

Несмотря на снижение среднего чека, финансовый результат эксперимента положительный за счёт кратного роста числа покупателей. Это особенно важно в условиях срочного падения выручки, когда  требуется быстрый эффект.
